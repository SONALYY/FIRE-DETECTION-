```python
from keras.models import load_model
from tensorflow.keras.utils import load_img, img_to_array
from keras.applications.inception_v3 import InceptionV3,preprocess_input
import numpy as np
import matplotlib.pyplot as plt
model=load_model('mymodel.h5')


#path for the image to see if it predicts correct class
path="/content/MyDrive/test/smoking-2346593__340.webp"
img =load_img(path,target_size=(300,300))

i= img_to_array(img)

i = preprocess_input(i)

input_arr=np.array([i])
input_arr.shape

pred=np.argmax(model.predict(input_arr))

if pred==0:
    print("The image is forest fire")
    print("This image is a major effect from fire")
elif pred==1:
  print("The image is Gas stove") 
  print("The image is a  medium effect of fire") 
elif pred==2:
  print("The image is Match Stick ") 
  print("The image is a minor effect of fire")
else:
  print("cannot predict")

#to display the image
plt.imshow(input_arr[0])
plt.title("input image")
plt.axis=False
plt.show()
```
